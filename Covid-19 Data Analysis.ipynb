This dataset is collected from Kaggle.com. This dataset contains latest Covid-19 India state-wise data as on August 25, 2021. This dataset can be used to analyze covid condition in India. 
In this notebook we are going to process and analyze the data using Numpy and Pandas libraries and then we will visualize it using matplotlib and seaborn library.

### Let's Import the modules

```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

## Importing Covid-19 dataset 

```python
data = pd.read_csv('Covid-19 India Status.csv')
```

## Visualizing the data 

```python
plt.figure(figsize=(15,5))
sns.barplot(x=data['State/UTs'],y=data['Total Cases'])
sns.set_style(style='darkgrid')
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(rotation=90)
plt.title('Total number of cases in all States')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x=data['State/UTs'],y=data['Active'])
sns.set_style(style='darkgrid')
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(rotation=90)
plt.title('Total number of active cases in all States')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x=data['State/UTs'],y=data['Discharged'])
sns.set_style(style='darkgrid')
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(rotation=90)
plt.title('Total number of discharged cases in all States')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x=data['State/UTs'],y=data['Deaths'])
sns.set_style(style='darkgrid')
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(rotation=90)
plt.title('Total number of deaths in all States')
plt.show()
```

```python
totcase_df = data.sort_values(by = 'Total Cases', ascending=False).head(10)
totcase_df.reset_index(inplace=True)
totcase_df.drop('index', axis=1, inplace=True)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='State/UTs', y='Total Cases', data=totcase_df)
sns.set_style(style='darkgrid')
plt.title('Top 10 states with highest number of cases')
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```

```python
actcase_df = data.sort_values(by='Active', ascending=False).head(10)
actcase_df.reset_index(inplace=True)
actcase_df.drop('index', axis=1, inplace=True)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='State/UTs', y='Active', data=actcase_df)
sns.set_style(style='darkgrid')
plt.title("Top 10 states with highest number of active cases")
plt.show()
```

```python
discase_df = data.sort_values(by='Discharged', ascending=False).head(10)
discase_df.reset_index(inplace=True)
discase_df.drop('index', axis=1, inplace=True)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='State/UTs', y='Discharged', data=discase_df)
sns.set_style(style='darkgrid')
plt.ticklabel_format(style='plain', axis='y')
plt.title('Top 10 states with highest number of discharged cases')
plt.show()
```

```python
deathcase_df = data.sort_values(by = 'Deaths', ascending=False).head(10)
deathcase_df.reset_index(inplace=True)
deathcase_df.drop('index', axis=1, inplace=True)
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='State/UTs',y='Deaths', data=deathcase_df)
sns.set_style(style='darkgrid')
plt.title('Top 10 states with highest number of deaths')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(x=data['Total Cases'],y=data['Active'])
sns.set_style(style='darkgrid')
plt.ticklabel_format(style='plain', axis='x')
plt.title('Total Cases vs Active Cases')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(x=data['Total Cases'], y=data['Deaths'])
sns.set_style(style='darkgrid')
plt.ticklabel_format(style='plain', axis='x')
plt.title('Total Cases vs No. of Deaths')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(x=data['Total Cases'], y=data['Discharged'])
sns.set_style(style='darkgrid')
plt.ticklabel_format(style='plain', axis='both')
plt.title('Total Cases vs Discharged Cases')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.lineplot(x=data['Active'], y=data['Deaths'])
sns.set_style(style='darkgrid')
plt.title('Active Cases vs No. of Deaths')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='State/UTs', y='Active Ratio (%)', data=data)
sns.set_style(style='darkgrid')
plt.xticks(rotation=90)
plt.title('Active Cases Ratio')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='State/UTs', y='Discharge Ratio (%)', data=data)
sns.set_style(style='darkgrid')
plt.xticks(rotation=90)
plt.title('Discharge Ratio')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(x='State/UTs', y='Death Ratio (%)', data=data)
sns.set_style(style='darkgrid')
plt.xticks(rotation=90)
plt.title('Death Ratio')
plt.show()
```
